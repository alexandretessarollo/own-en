We'll use the defsATWNlist_clear as the main reference of "id".

For a given id:
- if both AT and WN share the same id
    - any common rels will be disregarded (they'll be deemed to be already in Freeling files)
    - rels only in WN will go to a list to be removed (we need to build the w-rel tuplets later)
    - rels only in AT will go to a list to be added (we need to build the w-rel tuplets later)
- if id is only in AT, all links will go to a list to be added (we need to build the w-rel tuplets later)
- if id is only in WN, all links will go to a list to be removed (we need to build the w-rel tuplets later)

Now lets go through defsATWNlist_clear and, for each item that shows ids for both AT and WN, go into relsAT and relsWN and check whether the relations are common (and remove them) or not (and keep them, later we'll consolidate it).

```python
# I don't know why, but just one code run won't do the trick. Albeit not elegant, the loop introduced with the while does the trick. Apparently it takes 9 runs to exhaust the relations - probably something to do with three levels of for...
# On a second thought, it might be because I'm changing the lists the fors are running on, so I'm likely messing everything. It could be solved with two temporary lists (so 2 lists are used on the fors and the other 2 on the editing).
a = 1
b = 1
counter = 0

while a != len(relsAT) and b != len(relsWN):  
    a = len(relsAT)
    b = len(relsWN)
    for n, s in enumerate(defsATWNlist_clear):
        if s[0] == -1 or s[1] == -1:
            continue
        else:
            for m,r in enumerate(relsAT):
                if r[0] == s[0]: #same ATid
                    for p,q in enumerate(relsWN):
                        if q[0] == s[1]: #same WNid
                            if r[1] == q[1]:
                                relsWN.remove(q)
                                relsAT.remove(r)
                                break
                            else:
                                continue
                        else:
                            continue
                else:
                    continue
    counter = counter + 1
    print(counter)
```

# Yey! Now we know which relations must go (relsWN) and which must be added (relsAT)

Notice relsAT is a list of doubles in the format [idAT,w2AT]. A proper relation is always composed of [w1AT, w2AT]. We'll need to replace idAT with w1AT, and we'll use wordsAT for that. Synonyms (multiple w:'s for one given d:) we'll generate several pairs of relation for the same second-part of such relation ([w1AT_synonym1, w2AT], [w1AT_synonym2, w2AT], ...). The same goes for relsWN.

But before we venture into this aspect, let's do the same "cleaning" in the words lists

```python
# I don't know why, but just one code run won't do the trick. Albeit not elegant, the loop introduced with the while does the trick. Apparently it takes 9 runs to exhaust the relations - probably something to do with three levels of for...
a = 1
b = 1
counter = 0

while a != len(wordsAT) and b != len(wordsWN):  
    a = len(wordsAT)
    b = len(wordsWN)
    for n, s in enumerate(defsATWNlist_clear):
        if s[0] == -1 or s[1] == -1:
            continue
        else:
            for m,r in enumerate(wordsAT):
                if r == s[0]: #same ATid
                    for p,q in enumerate(wordsWN):
                        if q == s[1]: #same WNid
                            if wordsAT[m+1] == wordsWN[p+1]:
                                del wordsAT[m+1]
                                del wordsAT[m]
                                del wordsWN[p+1]
                                del wordsWN[p]
                                break
                            else:
                                continue
                        else:
                            continue
                else:
                    continue
    counter = counter + 1
    print(counter)
```
