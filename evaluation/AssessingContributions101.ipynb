```python
import nltk, re, pprint
from nltk import word_tokenize
from urllib import request
```

```python
url = 'https://raw.githubusercontent.com/alexandretessarollo/own-en/master/src/noun.substance'
```

```python
response = request.urlopen(url)
raw = response.read().decode('utf8')
```

```python
subsAT = nltk.line_tokenize(raw)
```

```python
for n,s in enumerate(subsAT): #aim: to change every 'w: example drf etc' to 'w: example'
    if s.find("w: ") == -1: #if it is not a w: move on
        next
    elif s.find(" ", 3) == -1: # if it is a w: but already correct (w: example), move on
        next
    else: # replace 'w: example drf etc' with 'w: example'
        subsAT[n]=s[:s.find(" ", 3)]
```

```python
tracker = 0
sense_id = 1
for n,s in enumerate(subsAT):
    if tracker == 0:
        subsAT[n]=[sense_id,subsAT[n]]
        if s.find("d: ") == 0:
            tracker = 1
        continue
    elif s.find("w: ") == 0:
        sense_id = sense_id+1
        subsAT[n]=[sense_id,subsAT[n]]
        tracker = 0
    else:
        subsAT[n]=[sense_id,subsAT[n]]
```

```python
wordsAT = []
for n,s in enumerate(subsAT):
    if s[1].find("w: ") == 0:
        wordsAT = wordsAT + s
wordsAT
```
